```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
from ultralytics import YOLO
```

```python
# Load YOLOv8 model (pre-trained)
model = YOLO("yolov8n.pt")
```

```python
def detect_vehicles(frame):
    """
    Detect vehicles in a frame using YOLOv8.
    Returns bounding boxes [(x1, y1, x2, y2, class_id, color)].
    """
    results = model(frame)
    vehicles = []

    for r in results:
        for box in r.boxes:
            x1, y1, x2, y2 = box.xyxy[0].tolist()
            class_id = int(box.cls[0])

            # Consider only vehicle classes (Car, Truck, Bus, Motorcycle)
            if class_id in [2, 3, 5, 7]:
                color = get_vehicle_color(frame, (x1, y1, x2, y2))  # Extract vehicle color
                vehicles.append((x1, y1, x2, y2, class_id, color))  # Return 6 values

    print(f"\nüü¢ [DETECTION] Vehicles Detected: {len(vehicles)}")
    return vehicles
```

```python
def get_vehicle_color(frame, bbox):
    """
    Extracts the dominant color from the vehicle bounding box.
    Returns the mean (R, G, B) color.
    """
    x1, y1, x2, y2 = map(int, bbox)
    vehicle_roi = frame[y1:y2, x1:x2]

    if vehicle_roi.size == 0:  # Prevent errors with empty ROIs
        return (0, 0, 0)

    mean_color = cv2.mean(vehicle_roi)[:3]  # (B, G, R)
    return (int(mean_color[2]), int(mean_color[1]), int(mean_color[0]))  # Convert to (R, G, B)
```

```python
class VehicleTracker:
    """
    Tracks vehicles using type, color, centroid distance, and motion.
    """
    def __init__(self):
        self.vehicles = {}  # {vehicle_id: (last_frame, last_centroid, direction, class_id, color)}
        self.next_vehicle_id = 1  # Unique ID counter

    def update(self, detected_vehicles, frame_idx, frame):
        """
        Updates the tracker, classifies vehicles, and draws bounding boxes.
        Returns new, old vehicles and frame with visualization.
        """
        print(f"üîµ [FRAME {frame_idx}] Vehicles Detected: {len(detected_vehicles)}")

        new_vehicles = []
        old_vehicles = []
        frame_with_boxes = frame.copy()
        updated_vehicle_ids = set()

        for (x1, y1, x2, y2, class_id, color) in detected_vehicles:
            centroid = ((x1 + x2) / 2, (y1 + y2) / 2)
            assigned_vehicle_id = None
            min_distance = float("inf")

            # Try to match detected vehicle with an existing one
            for vid, (last_frame, last_centroid, direction, last_class, last_color) in self.vehicles.items():
                if last_class != class_id:  # Ensure type consistency
                    continue

                # Compute Euclidean distance
                distance = np.sqrt((centroid[0] - last_centroid[0])**2 + (centroid[1] - last_centroid[1])**2)

                # Compute color similarity
                color_diff = np.linalg.norm(np.array(last_color) - np.array(color))

                # Match condition: distance, color similarity, and time consistency
                if distance < 50 and color_diff < 50 and frame_idx - last_frame <= 3:
                    if distance < min_distance:  # Choose the closest match
                        min_distance = distance
                        assigned_vehicle_id = vid

            # If a match is found, update the vehicle
            if assigned_vehicle_id is not None:
                new_direction = "Incoming" if centroid[1] > self.vehicles[assigned_vehicle_id][1][1] else "Outgoing"
                self.vehicles[assigned_vehicle_id] = (frame_idx, centroid, new_direction, class_id, color)
                old_vehicles.append((assigned_vehicle_id, new_direction))
                updated_vehicle_ids.add(assigned_vehicle_id)
                print(f"    üîπ Vehicle {assigned_vehicle_id} updated: {new_direction}")

            # If no match was found, assign a new ID
            if assigned_vehicle_id is None:
                direction = "Incoming" if centroid[1] > frame.shape[0] / 2 else "Outgoing"
                self.vehicles[self.next_vehicle_id] = (frame_idx, centroid, direction, class_id, color)
                new_vehicles.append((self.next_vehicle_id, direction))
                assigned_vehicle_id = self.next_vehicle_id
                self.next_vehicle_id += 1
                print(f"    üÜï [NEW VEHICLE] ID: {assigned_vehicle_id}, Direction: {direction}")

            # Draw bounding box and ID
            cv2.rectangle(frame_with_boxes, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            if(direction=="Outgoing"):
              cv2.putText(frame_with_boxes, f"{assigned_vehicle_id}-Og", (int(x1), int(y1) - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0, 255, 0), 1)
            else:
              cv2.putText(frame_with_boxes, f"{assigned_vehicle_id}-In", (int(x1), int(y1) - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0, 255, 0), 1)


        return new_vehicles, old_vehicles, frame_with_boxes
```

```python
def process_video(input_video_path, output_video_path, output_csv_path, start_time_sec, end_time_sec):
    """
    Process a video, track unique vehicles in the given time range,
    classify them as Incoming or Outgoing, and save results.
    """
    cap = cv2.VideoCapture(input_video_path)

    # Get video properties
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    total_duration = frame_count / fps  # Total video duration in seconds

    # Convert input times to frame indices
    start_frame = int(start_time_sec * fps)
    end_frame = int(end_time_sec * fps)

    if end_time_sec > total_duration:
        print("\n‚ö† [WARNING] End time exceeds video duration! Using last frame instead.")
        end_frame = frame_count - 1

    # Initialize vehicle tracker
    tracker = VehicleTracker()

    # Set up video writer for output
    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
    out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

    frame_idx = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret or frame_idx > end_frame:
            break

        # Process only frames in the selected interval
        if frame_idx >= start_frame:
            # Detect vehicles
            vehicles = detect_vehicles(frame)

            # Update tracker
            new_vehicles, old_vehicles, frame_with_boxes = tracker.update(vehicles, frame_idx, frame)

            # Print debug statements **only at every full second**
            if frame_idx % fps == 0:
                second_elapsed = frame_idx // fps
                print("\nüü† [TIME UPDATE] -------------------------------------")
                print(f"üïí Time Elapsed: {second_elapsed} sec")
                print(f"üìä Vehicles Detected: {len(vehicles)} in this second")
                print("üöó New Vehicles:")
                for vid, direction in new_vehicles:
                    print(f"    - Vehicle {vid}: {direction}")
                print("üöô Old Vehicles:")
                for vid, direction in old_vehicles:
                    print(f"    - Vehicle {vid}: {direction}")
                print("--------------------------------------------------")

                # Display the frame with bounding boxes using Matplotlib
                plt.figure(figsize=(10, 6))
                plt.imshow(cv2.cvtColor(frame_with_boxes, cv2.COLOR_BGR2RGB))
                plt.title(f"Frame {frame_idx} - Time: {second_elapsed} sec")
                plt.axis('off')
                plt.show()


            out.write(frame_with_boxes)  # Save processed frame

        frame_idx += 1

    cap.release()
    out.release()
    cv2.destroyAllWindows()  # Close all OpenCV windows after processing

    # Final results
    result = {
        "Incoming Vehicles": len([v for v in tracker.vehicles.values() if v[2] == "Incoming"]),
        "Outgoing Vehicles": len([v for v in tracker.vehicles.values() if v[2] == "Outgoing"])
    }

    # Save to CSV
    pd.DataFrame([result]).to_csv(output_csv_path, index=False)

    print("\n‚úÖ [PROCESS COMPLETE] -----------------------------------")
    print(f"üìå Processed Video Interval: {start_time_sec} - {end_time_sec} sec")
    print(f"üìÇ Vehicle Counts Saved to: {output_csv_path}")
    print(f"üöó Incoming Vehicles: {result['Incoming Vehicles']}")
    print(f"üöô Outgoing Vehicles: {result['Outgoing Vehicles']}")
    print("---------------------------------------------------------")
```

```python
input_video = "./frames/output739864242.mp4"
output_video = "./result/output_lane_detected.mp4"
output_csv = "./result/vehicle_counts.csv"

# Define the time range (e.g., analyze vehicles from 10s to 30s)
start_time_sec = 15
end_time_sec = 20

process_video(input_video, output_video, output_csv, start_time_sec, end_time_sec)
```
